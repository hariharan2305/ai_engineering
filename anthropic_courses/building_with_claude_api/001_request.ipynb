````python
# Loading environment variables from a .env file
from dotenv import load_dotenv
load_dotenv()
````

````python
# Creating an API client
from anthropic import Anthropic

client = Anthropic()
model = "claude-haiku-4-5"
````

````python
def add_user_message(messages: list, content: str) -> None:
    """
    Helper function to add a user message to the messages list.
    """
    messages.append({
        "role": "user",
        "content": content
    })
    # return messages

def add_assistant_message(messages: list, content: str) -> None:
    """
    Helper function to add an assistant message to the messages list.
    """
    messages.append({
        "role": "assistant",
        "content": content
    })
    # return messages


def chat(messages: list, model: str = model, max_tokens: int = 1000) -> str:
    """
    Function to send a chat request to the Claude API and return the assistant's response.
    """
    response = client.messages.create(
        model=model,
        max_tokens=max_tokens,
        messages=messages
    )
    return response.content[0].text
````

````python
# Empty list to hold the conversation messages
messages = []

# Adding a user message to the conversation
add_user_message(messages, "What is quantum computing?")

# Sending the chat request and printing the assistant's response
response = chat(messages)

# Add the assistant's response to the messages list
add_assistant_message(messages, response)

# Ask a follow-up question
add_user_message(messages, "Can you give one interesting fact about this topic?")

# Get the assistant's response to the follow-up question
response = chat(messages)
print(response)
````

### Creating a basic chat app:

````python
# Empty list to hold the conversation messages
messages = []

# Adding a loop to have a conversation with the assistant
while True:
    user_message = input("> ")
    print("> ", user_message)
    add_user_message(messages, user_message)
    response = chat(messages)
    print("> ", response)
    add_assistant_message(messages, response)
    print("----------------------------------\n")
````

````python
# More efficient way to get the streaming responses both events and final fully generated final response
messages = []

add_user_message(messages, "Give me a one sentence movie idea")

# Use this to stream the response to the client for UI/UX
with client.messages.stream(
    model=model,
    max_tokens=1000,
    messages=messages
) as stream:
    for text in stream.text_stream:
        print(text, end="")

# Use this to get the final fully generated response after streaming
# final_response = stream.get_final_message() # returns the full message object with all metadata
final_response = stream.get_final_text()
print("\n\n\nFinal response:\n", final_response)
````

### Structured Data:

````python
def add_user_message(messages: list, content: str) -> None:
    """
    Helper function to add a user message to the messages list.
    """
    messages.append({
        "role": "user",
        "content": content
    })
    # return messages

def add_assistant_message(messages: list, content: str) -> None:
    """
    Helper function to add an assistant message to the messages list.
    """
    messages.append({
        "role": "assistant",
        "content": content
    })
    # return messages


def chat(messages: list, model: str = "claude-haiku-4-5", max_tokens: int = 1000, system_prompt: str = None, temperature: float = 0.0, stop_sequences: list = None) -> str:
    """
    Function to send a chat request to the Claude API and return the assistant's response.
    """
    params = {
        "model": model,
        "max_tokens": max_tokens,
        "messages": messages,
        "temperature": temperature,
        "stop_sequences": stop_sequences
    }
    if system_prompt:
        params["system"] = system_prompt

    response = client.messages.create(**params)
    return response.content[0].text
````

````python
messages = []

add_user_message(messages, "Generate numbers from 1 to 10")

chat(messages, stop_sequences=[ "5"])
````

````python
messages = []

add_user_message(messages, "Generate a single sample json response with name, age and city")

print("\n Response without tuning for structured response: ", chat(messages))

# Add assistant message to steer the model towards structured response
add_assistant_message(messages, "```json")

print("\n Response after tuning for structured response: ", chat(messages, stop_sequences=["```"]))
````
